```python
lack_list=[]
def get_img_from_imdb(movie_id:int):
    base_url='https://www.imdb.com/title/tt'
    url=base_url+"{:0>7d}".format(movie_id)
    header={
        'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:109.0) Gecko/20100101 Firefox/110.0'
    }
    response=requests.get(url=url,headers=header)
    assert(response.status_code==200)
    # print(response.status_code)
    tree=etree.HTML(response.text)
    img=tree.xpath('//div[@class="ipc-media ipc-media--poster-27x40 ipc-image-media-ratio--poster-27x40 ipc-media--baseAlt ipc-media--poster-l ipc-poster__poster-image ipc-media__img"]/img/@src')
    # assert(len(img)==1)
    if(len(img)==0):
        lack_list.append(movie_id)
        print("finished {},lack".format(movie_id))
        return
    assert(len(img)==1)
    save_path='./photo'
    # print(img)
    if(os.path.exists(save_path)==False):
        os.makedirs(save_path)
    # os.makedirs(save_path)
    img_name=save_path+'/'+str(movie_id)+'.jpg'
    img_data=requests.get(img[0])
    with open(img_name,'wb') as fp:
        fp.write(img_data.content)
    print("finished {},done".format(movie_id))
    # print(response.status_code)
```

```python
get_img_from_imdb(116583)
```
